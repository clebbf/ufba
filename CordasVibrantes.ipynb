```python
#Funções que realizam MMQ e geram Gráficos
#Código feito e revisad

#Libraries Utilizadas
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
from matplotlib.ticker import AutoMinorLocator, FormatStrFormatter
import numpy as np

#Função para calcular o MMQ
def funcaoMMQ(arrX, arrY,log):
    #MMQ no log-log
    if log == 'sim':
        logX = np.log(arrX)
        logY = np.log(arrY)
        XY = []
        X2 = []
        for i in range(len(logX)):
                XY.append(logX[i] * logY[i])
        for i in range(len(logX)):
                X2.append(logX[i] * logX[i])
        
        a = (5*(sum(XY)) - (sum(logX)*sum(logY)))/(5*(sum(X2)) - sum(logX)**2)
        b = (sum(logY) - a*sum(logX))/5
        
        Ya = []
        
        for i in range(len(logX)):
          y = a*logX[i] + b
          Ya.append(y)
    
        print('O somatorio de log(X):',sum(logX),"\n",'O somatorio de log(Y):',sum(logY),"\n",'O somatorio de log(X)*log(Y):',sum(XY),"\n",'O somatorio de log(X)^2:',sum(X2),"\n")
        print(f'O valor da constante a é {a}',"\n",f'O valor da constante b é {b}')

#MMQ no milimetrado
    elif log =='nao':
        XY = []
        X2 = []
        for i in range(len(arrX)):
                XY.append(arrX[i] * arrY[i])
        for i in range(len(arrX)):
                X2.append(arrX[i] * arrX[i])

        a = (5*(sum(XY)) - (sum(arrX)*sum(arrY)))/(5*(sum(X2)) - sum(arrX)*sum(arrX))
        b = (sum(arrY) - a*sum(arrX))/5

        Ya=[]

        print(arrX, arrY)

        for i in range(len(arrX)):
          y = a*arrX[i] + b
          Ya.append(y)

        print('O somatorio de log(X):',sum(arrX),"\n",'O somatorio de log(Y):',sum(arrY),"\n",'O somatorio de log(X)*log(Y):',sum(XY),"\n",'O somatorio de log(X)^2:',sum(X2),"\n")
        print(f'O valor da constante a é {a:,.2f}',"\n",f'O valor da constante b é {b:,.2f}')

#Função para criar o gráfico
def funcaoGrafico(arrX, arrY, nome, eixoX, eixoY, log):
    if log == 'nao':
        fig, ax = plt.subplots()
        plt.scatter(arrX, arrY, color='red')
        plt.title(nome)
        
        plt.xlabel(eixoX)
        plt.ylabel(eixoY)

        ax.set_xlim(arrX[0]-(0.50*arrX[0]), arrX[4]+(0.50*arrX[4]))
        ax.set_ylim(arrY[0]-(0.50*arrY[0]), arrY[4]+(0.50*arrY[4]))

        ax.yaxis.grid(True)
        ax.xaxis.grid(True)

        plt.xticks(arrX, fontsize=7)
        plt.yticks(arrY, fontsize=7)

        plt.grid(visible=True, which='major', linewidth=1, linestyle='--')
        plt.grid(visible=True, which='minor', linewidth=1, linestyle='--')

        plt.show()
    elif log == 'sim':
        fig, ax = plt.subplots(figsize=(12,8))
        plt.scatter(arrX, arrY, color='red')
        plt.title(nome)
        
        plt.xlabel(eixoX)
        plt.ylabel(eixoY)

        # ax.set_xlim(arrX[0]-(0.50*arrX[0]), arrX[4]+(0.50*arrX[4]))
        # ax.set_ylim(arrY[0]-(0.50*arrY[0]), arrY[4]+(0.50*arrY[4]))

        ax.yaxis.grid(True)
        ax.xaxis.grid(True)

        ax.set_xscale("log")
        ax.set_yscale("log")
        
        plt.xticks(arrX, fontsize=7)
        plt.yticks(arrY, fontsize=7)

        plt.grid(visible=True, which='major', linewidth=1, linestyle='--')
        plt.grid(visible=True, which='minor', linewidth=1, linestyle='--')

        plt.minorticks_off()
        plt.show()
    
```

```python
#Gráfico Frequência X Número de Nós
#MMQ Frequência X Número de Nós

n =[1, 2, 3, 4, 5]
freq1 = [22.33 , 42.13, 57.73, 72.53, 86.19]



funcaoMMQ(n, freq1, 'nao')
funcaoGrafico(n, freq1, 'Gráfico da Frequência X Número de Nós', 'Número de Nós', 'Frequência (Hz)', 'nao')
```

```python
#Gráfico Frequência X Comprimento
#MMQ Frequência X Comprimento

L = np.array([1.10, 1.20, 1.30, 1.40, 1.50])
freq2 = np.array([41.13 , 37.66, 34.53, 32.14, 29.40])

funcaoMMQ(L, freq2, 'sim')
funcaoGrafico(L, freq2, 'Gráfico da Frequência X Comprimento', 'Comprimento (m)', 'Frequência (Hz)', 'sim')
```

```python
#Gráfico Frequência X Tração
#MMQ Frequência X Tração

tracao = np.array([0.7154, 0.8114, 0.9074, 1.0035, 1.0995])
freq3 = np.array([45.33, 47.13, 49.33, 51.33, 53.73])

funcaoMMQ(tracao, freq3, 'sim')
funcaoGrafico(tracao, freq3, 'Gráfico da Frequência X Tração', 'Tração (N)', 'Frequência (Hz)', 'sim')
```

```python
#Gráfico Frequência X Densidade Linear
#MMQ Frequência X Densidade Linear

dLinear= np.array([0.34, 0.507, 0.620, 0.787, 1.001])
freq4 = np.array([54.08, 47.43, 44.26, 41.80, 38.13])

#Correção de grama/metro -> kilograma/metro
for i in range(len(dLinear)):
    dLinear[i] = dLinear[i]/1000
    
funcaoMMQ(dLinear, freq4, 'sim')
funcaoGrafico(dLinear, freq4, 'Gráfico da Frequência X Densidade Linear', 'Densidade Linear (kg/m)', 'Frequência (Hz)', 'sim')
```
